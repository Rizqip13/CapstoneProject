### My Func

```python
def infotable(x):
    return pd.DataFrame(columns='dataFeatures dataType null nullPct unique uniqueSample'.split(),
                        data=[[col, 
                               x.dtypes[col], 
                               sum(x[col].isna()),
                               round(100*sum(x[col].isna())/len(x[col]),2), 
                               x[col].nunique(),list(x[col].unique()[:5])]
                              for col in x.columns])       

def variabletypes(dataType, unique):
    if 'object' == dataType:
        return 'Categorical'
    elif ('int'in str(dataType)) & (unique == 2):
        return 'Binary'
    else :
        return 'Numerical'
```

```python
dfDescribe= infotable(df)
pd.options.display.max_rows = len(dfDescribe)
dfDescribe
```

```python
print('columns with only one nunique ',dfDescribe[dfDescribe['unique']==1]['dataFeatures'].values)
```

```python
data=df.drop(dfDescribe[dfDescribe['unique']==1]['dataFeatures'].values,axis=1)
```

```python
dataDescribe = infotable(data)
dataDescribe['Variable type']=dataDescribe.apply(lambda row: variabletypes(row['dataType'], row['unique']),axis=1)
dataDescribe
```

```python
for x in dataDescribe['Variable type'].unique():
    print (str(sum(dataDescribe['Variable type']==x)).rjust(3),x.rjust(14),'\t: ',dataDescribe[dataDescribe['Variable type']==x]['dataFeatures'].values)
```

## When you look at the categorical variables, some have more than 10 nunique

```python
import category_encoders as ce
```

```python
cat_data=data.iloc[:,2:10].copy()
cat_data.head()
```

```python
encoder = ce.BinaryEncoder(cols=list(cat_data.columns), drop_invariant=True)
encoder
```

```python
X = encoder.fit_transform(X)
X.head()
```

```python
y.head()
```

```python
from xgboost import XGBRegressor, plot_importance
```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=101)
```

```python
XGBmodel = XGBRegressor(random_state=101)
XGBmodel
```

```python
XGBmodel.fit(X_train,y_train)
trainpred=XGBmodel.predict(X_train)
predictions=XGBmodel.predict(X_test)
```

```python
plt.scatter(y_train,trainpred)
plt.title('train predictions')
```

```python
plt.scatter(y_test,predictions)
plt.title('test predictions')
```

```python
sns.distplot((y_test['y'].values-predictions), bins=50)
```

```python
print('MAE : ', mean_absolute_error(y_test, predictions))
print('MSE : ', mean_squared_error(y_test, predictions))
print('RMSE : ', np.sqrt(mean_squared_error(y_test, predictions)))
print('R2 Score : ', r2_score(y_test, predictions))
```
